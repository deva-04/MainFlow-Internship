```python
!pip install pandas
!pip install seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:\\Users\\banda\\Downloads\\householdtask3.csv"
data = pd.read_csv(file_path, encoding='utf-8')

plt.figure(figsize=(10, 6))

# Scatter Plot
plt.subplot(221)
sns.scatterplot(x='year', y='own', data=data)
plt.title("Scatter Plot")
plt.xlabel("Year")
plt.ylabel("Ownership")

# Line Chart
plt.subplot(222)
sns.lineplot(x='year', y='own', data=data)
plt.title("Line Chart")
plt.xlabel("Year")
plt.ylabel("Ownership")

# Bar Chart
plt.subplot(223)
sns.barplot(x='year', y='own', data=data)
plt.title("Bar Chart")
plt.xlabel("Year")
plt.ylabel("Ownership")

# Histogram
plt.subplot(224)
sns.histplot(data['income'], kde=True)
plt.title("Income Histogram")

plt.tight_layout()

plt.show()
```
