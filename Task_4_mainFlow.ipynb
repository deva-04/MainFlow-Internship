## Perform exploratory data analysis (EDA) on a dataset. Create visualizations to understand the distribution of variables, identify outliers, and check for correlation between variables.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

#data = pd.read_csv('USvideos.csv')
data = pd.read_csv("C:\\Users\\banda\\Downloads\\USvideos.csv")

print("First few rows of the dataset:\n")
print(data.head(), "\n")

print("Shape of the dataset:\n")
print(data.shape, "\n")

print("Information about the dataset:\n")
print(data.info(), "\n")

print("Summary statistics of the dataset:\n")
print(data.describe(), "\n")

print("Columns in the dataset:\n")
print(data.columns.tolist(), "\n")

print("Count of missing values in the dataset:\n")
print(data.isnull().sum(), "\n")

print("Number of unique values in each column:\n")
print(data.nunique(), "\n")

# Plot count plot of 'category_id'
print("Count Plot of Category ID:\n")
plt.figure(figsize=(12, 8))
sns.countplot(x='category_id', data=data, palette='viridis')
plt.title('Count Plot of Category ID')
plt.xlabel('Category ID')
plt.ylabel('Count')
plt.show()

# Plot distribution of numerical features: views, likes, dislikes, comment_count
print("Distribution of Numerical Features:\n")
plt.figure(figsize=(14, 10))
numerical_features = ['views', 'likes', 'dislikes', 'comment_count']
for i, col in enumerate(numerical_features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel('')
plt.suptitle('Distribution of Numerical Features', y=1.02)
plt.tight_layout()
plt.show()

# Plot strip plot for category_id and views
print("Strip Plot for Category ID and Views:\n")
plt.figure(figsize=(14, 10))
sns.stripplot(x="category_id", y="views", data=data, palette='viridis', size=3, jitter=True)
plt.title('Strip Plot for Category ID and Views')
plt.xlabel('Category ID')
plt.ylabel('Views')
plt.show()

# Plot pair plot of numerical features
print("Pair Plot for Numerical Features:\n")
sns.pairplot(data[numerical_features], palette='Pastel1')
plt.suptitle('Pair Plot for Numerical Features', y=1.02)
plt.show()

# Plot violin plot for category_id and likes
print("Violin Plot for Category ID and Likes:\n")
plt.figure(figsize=(14, 10))
sns.violinplot(x="category_id", y="likes", data=data, palette='Set2', alpha=0.7)
plt.title('Violin Plot for Category ID and Likes')
plt.xlabel('Category ID')
plt.ylabel('Likes')
plt.show()

# Plot box plot between category_id and comment_count
print("Box Plot for Category ID and Comment Count:\n")
plt.figure(figsize=(14, 10))
sns.boxplot(x='category_id', y='comment_count', data=data, palette='Set3')
plt.title('Box Plot for Category ID and Comment Count')
plt.xlabel('Category ID')
plt.ylabel('Comment Count')
plt.show()

# Plot correlation heatmap for numerical features
print("Correlation Heatmap:\n")
plt.figure(figsize=(14, 10))
correlation_matrix = data[numerical_features].corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='Pastel2', linewidths=2)
plt.title('Correlation Heatmap')
plt.show()
```
